## **SCRAPPING : NUMBEO_COST_OF_LIVING**

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.numbeo.com/cost-of-living/rankings.jsp?title=2024'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
soup.find('table', class_ = 'stripe row-border order-column compact')
soup.find_all('table')[1]
table = soup.find_all('table')[1]
print(table)
soup.find_all('th')
Cost_of_living_titles = soup.find_all('th')
Cost_of_living_table_titles = [title.text for title in Cost_of_living_titles]
print(Cost_of_living_table_titles)
```

```python
import pandas as pd
df = pd.DataFrame (columns = Cost_of_living_table_titles)
df
```

```python
columns_data = soup.find_all('tr')
```

```python
for row in columns_data [2:372] :
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = individual_row_data

for i in range(0,len(df)) :
    df.loc[i,"Rank"] = i+1
```

```python
df
```

```python
# Chemin vers le dossier Move_Smart dans sample_data
df.to_csv('/content/sample_data/Move_Smart/Cost_of_living.csv', index=False)
```

## **MERGE ALL THE NUMBEO DATA**



```python
import pandas as pd

# Liste des fichiers CSV  charger
files = ['Affordability.csv', 'Cost_of_living.csv', 'Crime_Safety.csv', 'Health_Care.csv',
         'Pollution.csv', 'Quality_of_life.csv', 'Traffic.csv']

# Lire le fichier "Cost_of_living.csv" qui contient le plus grand nombre de villes
merged_df = pd.read_csv(f'/content/sample_data/Move_Smart/{files[1]}')  # Cost_of_living.csv

# Fusionner les autres fichiers sur la colonne "City" en utilisant "left" join pour conserver toutes les villes du fichier "Cost_of_living.csv"
for file in files[0:1] + files[2:]:  # Exclure le fichier "Cost_of_living.csv" de la liste pour viter de le fusionner avec lui-mme
    df = pd.read_csv(f'/content/sample_data/Move_Smart/{file}')

    # Fusionner les DataFrames en ajoutant des suffixes pour les colonnes en conflit
    merged_df = pd.merge(merged_df, df, on="City", how="left", suffixes=('', '_dup'))

# Si vous voulez supprimer les colonnes en conflit qui ont le suffixe '_dup'
merged_df = merged_df.loc[:, ~merged_df.columns.str.endswith('_dup')]

# Exporter le DataFrame fusionn dans un fichier CSV avec le nouveau nom
merged_df.to_csv('/content/sample_data/Move_Smart/all_city_data_merged_move_smart.csv', index=False)

print("Fusion termine et exporte dans 'all_city_data_merged_move_smart.csv'")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

indices = [ 'cost_of_living_index', 'rent_index', 'groceries_index',
       'restaurant_price_index', 'local_purchasing_power_index',
       'quality_of_life_index', 'health_care_index', 'crime_index',
       'pollution_index', 'traffic_index_(inneficiency)', 'safety_index',
        'climate_index', 'time_index(in_minutes)',
       '%racial_acceptance', '%_lgbtq_acceptance', 'freedom_score' ]

# Loop through each index to generate histogram and basic statistics
for index in indices:
    # Drop any missing data (NaN values) from the column
    data = df[index].dropna()


# Plot histogram
    plt.figure(figsize=(10, 6))
    sns.histplot(data, kde=True, bins=20, color='skyblue')
    plt.title(f"Distribution of {index}")
    plt.xlabel(index)
    plt.ylabel("Frequency")

  # Calculate and print mean, median, and mode
    mean = round(data.mean(), 2)
    median = round(data.median(), 2)
    mode = round(data.mode()[0], 2)

# Display mean, median, and mode on the plot
    plt.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean}')
    plt.axvline(median, color='green', linestyle='-', label=f'Median: {median}')
    plt.axvline(mode, color='blue', linestyle='-', label=f'Mode: {mode}')

    plt.legend()
    plt.show()

# Print basic statistics for each index
print(f"{index} - Mean: {mean:.2f}, Median: {median:.2f}, Mode: {mode:.2f}")
```
